## Challenge 1

```python
%%sql
SELECT tls.title AS TITLE, tls.title_id AS 'TITLE ID', au.au_id AS 'AUTHOR ID', tls.price * sls.qty * tls.royalty / 100 * ta.royaltyper / 100 AS 'SALES ROYALTY'
FROM authors au
JOIN 
    titleauthor ta ON au.au_id=ta.au_id
JOIN
    titles tls ON ta.title_id = tls.title_id  
JOIN
    sales sls ON tls.title_id = sls.title_id


    
```

```python
%%sql
SELECT tls.title AS TITLE, tls.title_id AS 'TITLE ID', au.au_id AS 'AUTHOR ID', SUM(tls.price * sls.qty * tls.royalty / 100 * ta.royaltyper / 100) AS 'AGGREGATE SALES ROYALTY'
FROM authors au
JOIN 
    titleauthor ta ON au.au_id=ta.au_id
JOIN
    titles tls ON ta.title_id = tls.title_id  
JOIN
    sales sls ON tls.title_id = sls.title_id
GROUP BY
    tls.title_id, au.au_id
```

```python
%%sql

SELECT au.au_id AS 'AUTHOR ID',  SUM(tls.advance) - SUM(tls.price * sls.qty * tls.royalty / 100 * ta.royaltyper / 100)  AS 'PROFITS'


FROM authors au
JOIN 
    titleauthor ta ON au.au_id=ta.au_id
JOIN
    titles tls ON ta.title_id = tls.title_id  
JOIN
    sales sls ON tls.title_id = sls.title_id
GROUP BY
    tls.title_id, au.au_id
    
ORDER BY 'PROFITS' desc
limit 3




```

```python
%%sql

SELECT au.au_id AS 'AUTHOR ID',  SUM(tls.advance) + SUM(tls.price * sls.qty * tls.royalty / 100 * ta.royaltyper / 100) AS 'PROFITS'


FROM authors au
LEFT JOIN 
    titleauthor ta ON au.au_id=ta.au_id
LEFT JOIN
    titles tls ON ta.title_id = tls.title_id  
LEFT JOIN
    sales sls ON tls.title_id = sls.title_id
GROUP BY
     au.au_id
    
ORDER BY 'PROFITS' desc 
limit 3
```

## Challenge 2

```python
%%sql

CREATE TABLE most_profiting_authors AS

SELECT 'AUTHOR ID' as au_id, PROFITS as profits

FROM

(SELECT au.au_id AS 'AUTHOR ID',  SUM(tls.advance) + SUM(tls.price * sls.qty * tls.royalty / 100 * ta.royaltyper / 100) AS 'PROFITS'


FROM authors au
LEFT JOIN 
    titleauthor ta ON au.au_id=ta.au_id
LEFT JOIN
    titles tls ON ta.title_id = tls.title_id  
LEFT JOIN
    sales sls ON tls.title_id = sls.title_id
GROUP BY
     au.au_id
    
ORDER BY 'PROFITS' desc)
```
